```python
def CreateListData(keywordslist):
    allitems = os.listdir('/data/ortho/')
    filteredlist = []
    for i in keywordslist:
        for x in allitems:
            if (i in x):
                filteredlist.append(x)
        allitems = filteredlist
        filteredlist = []
    finallist = []
    for i in allitems:
        finallist.append('/data/ortho/%s' % i)
    return finallist

AlleOefeningen = CreateListData(['meting'])
for num, i in enumerate(AlleOefeningen):
    print(num + 1, i)
```

```python
def CreateList(keywordslist):
    allitems = os.listdir('/datc/ortho/')
    filteredlist = []
    for i in keywordslist:
        for x in allitems:
            if (i in x):
                filteredlist.append(x)
        allitems = filteredlist
        filteredlist = []
    finallist = []
    for i in allitems:
        finallist.append('/datc/ortho/%s' % i)
    return finallist

AlleOefeningenC = CreateList(['meting'])
for num, i in enumerate(AlleOefeningenC):
    print(num + 1, i)
    
DoneList = []
for i in AlleOefeningen:
    name = i.split('/')[3]
    if name  not in [x.split('/')[3] for x in AlleOefeningenC]:
        DoneList.append(name)
```

```python
for num, i in enumerate(DoneList):
    print(num, i)
    
```
