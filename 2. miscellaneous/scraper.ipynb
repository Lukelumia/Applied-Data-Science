```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

```python
standardurl = "https://mcgill.ca/study/2018-2019/courses/search?f%5B0%5D=level%3Aundergraduate"
```

```python
data = {'name': [], 'faculteit': [], 'subject': [], 'level': [], 'periode': [], 'link': []}
for i in range(330):
    url = standardurl + '&page=' + str(i)
    
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page)
    
    table = soup.find("div", {"class": "view-content"})
    rows = table.find_all("div", {"class": "views-row"})
    
    
    for row in rows:
        links = row.find_all('a', href=True) 
        info = row.find_all('span', {"class": "field-content"})

        data['link'].append('https://mcgill.ca' + links[0]['href'])
        data['name'].append(links[0].text)
        data['faculteit'].append(info[0].text)
        data['subject'].append(info[1].text)
        data['level'].append(info[2].text)
        data['periode'].append(info[3].text)

    print('done with: %s' % i)
    
```

```python
df = pd.DataFrame.from_dict(data)
df.to_csv('vakken.csv', sep=';')
```
